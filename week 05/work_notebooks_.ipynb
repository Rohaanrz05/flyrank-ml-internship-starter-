**Loading** **Dataset**

```python
import pandas as pd

# Modified the URL to be a direct download link
file_id = "1bSpacQA05GVRYMiWGx-NeP_PDz6bEqgR"
url = f"https://drive.google.com/uc?export=download&id={file_id}"

df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

**Identifying Target Variable**

```python
target = "Investment_Recommendation"
y = df[target]
X = df.drop(target, axis=1)
print(X.shape)
print(y.shape)
```

**Handling Categorical Columns**

```python
X.dtypes
```

```python
X = pd.get_dummies(X, drop_first=True)
```

**Checking Missing Values**

```python
X.isnull().sum()
```

```python
X = X.fillna(X.median())
```

**Creat Training/Test Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

**Building Baseline Model**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

# Encode categorical target labels to numbers
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

# Use a classifier instead of Linear Regression
baseline = LogisticRegression(max_iter=1000)

# Fit the model
baseline.fit(X_train, y_train_encoded)

print(f"Model Score: {baseline.score(X_test, y_test_encoded):.4f}")
```

**Making Predictions**

```python
baseline_pred = baseline.predict(X_test)
```

**Evaluating Baseline**

```python
from sklearn.metrics import accuracy_score, classification_report

# Using classification metrics for the baseline model
baseline_acc = accuracy_score(y_test_encoded, baseline_pred)
print(f"Baseline Accuracy: {baseline_acc:.4f}")
print("\nClassification Report:\n", classification_report(y_test_encoded, baseline_pred, target_names=le.classes_))
```

**Training a Better Model**

```python
from sklearn.ensemble import RandomForestClassifier

# Use RandomForestClassifier instead of Regressor
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

# Fit using the encoded numerical labels
rf.fit(X_train, y_train_encoded)
```

***Predict Using Random Forest***

```python
rf_pred = rf.predict(X_test)
```

***Evaluating Random Forest***

```python
from sklearn.metrics import accuracy_score, classification_report

# Calculate accuracy for Random Forest
rf_acc = accuracy_score(y_test_encoded, rf_pred)
print(f"Random Forest Accuracy: {rf_acc:.4f}")
print("\nClassification Report:\n", classification_report(y_test_encoded, rf_pred, target_names=le.classes_))
```

***Comparing Models***

```python
results = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest"],
    "Accuracy": [baseline_acc, rf_acc]
})

results
```

***Feature Importance***

```python
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

importance.head(10)
```

***Visualize Importance***

```python
import matplotlib.pyplot as plt

importance.head(10).plot(
    x="Feature",
    y="Importance",
    kind="barh"
)

plt.show()
```

***Error Analysis***

```python
# Identify where predictions match the actual labels
correct_predictions = (y_test_encoded == rf_pred)
```

```python
comparison = pd.DataFrame({
    "Actual": le.inverse_transform(y_test_encoded),
    "Predicted": le.inverse_transform(rf_pred),
    "Is_Correct": correct_predictions
})

comparison.head(20)
```
