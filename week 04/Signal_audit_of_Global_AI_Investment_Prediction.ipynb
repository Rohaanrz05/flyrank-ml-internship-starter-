```python
import pandas as pd

# Adding compression='gzip' to handle the compressed file format
df = pd.read_csv("https://drive.google.com/uc?export=download&id=1rCSEY2B7hrKvDMypntgPoxRHrBRYhZ2B", compression='gzip')
df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.info()

df.describe()
```

```python
import matplotlib.pyplot as plt

df["AI_Investment_Decision"].value_counts().plot(kind="bar")

plt.title("Investment Decision Distribution")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(
    x="AI_Investment_Decision",
    y="Liquidity_Score",
    data=df
)

plt.title("Liquidity Score vs Investment Decision")
plt.show()
```

```python
df.groupby("AI_Investment_Decision")["Liquidity_Score"].mean()
```

```python
sns.boxplot(
    x="AI_Investment_Decision",
    y="Volatility_Index",
    data=df
)

plt.title("Volatility Index vs Decision")
plt.show()
df.groupby("AI_Investment_Decision")["Volatility_Index"].mean()
```

```python
sns.boxplot(
    x="AI_Investment_Decision",
    y="Social_Sentiment",
    data=df
)

plt.title("Social Sentiment vs Decision")
plt.show()
df.groupby("AI_Investment_Decision")["Social_Sentiment"].mean()
```

```python
pd.crosstab(
    df["Market_Trend"],
    df["AI_Investment_Decision"]
)
```
